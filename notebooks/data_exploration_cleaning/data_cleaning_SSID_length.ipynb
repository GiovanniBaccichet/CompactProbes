## Import Data

```python
# Combined dataframe
combined_df_csv = config["DEFAULT"]["interim_path"] + "balanced_df_raw.csv"
```

```python
df = pd.read_csv(combined_df_csv, index_col=0)
```

```python
df
```

Subtract from `Length` the length of the SSID, in order to not consider it

```python
df["Length_NO_SSID"] = df.apply(
    lambda row: (
        row["Length"] - len(row["SSID"])
        if isinstance(row["SSID"], str)
        else row["Length"]
    ),
    axis=1,
)
```
